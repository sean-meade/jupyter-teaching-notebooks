## Beginning the game

Below is some Python code to generate a random choice between three doors as pictured above: yellow, red and blue.
The code will randomly choose one of the doors to put the car behind.
There's no cheating here - the car is as likely to be behind any one of the doors.

The colour of the door with the car is stored as a string in the variable `car`.
The yellow door is indicated by the string `'yellow'`, the red by the string `'red'` and the blue by `'blue'`.
Please don't look at the value of the `car` variable, at least until the game is over.
If you re-run the cell you will again randomly select a door, possibly the same, possibly different.

````python
# Python provides a library called random to generate pseudo-random numbers and do stuff with them.
import random

# The three doors in a list.
doors = ['yellow', 'red', 'blue']

# Pick a random door.
car = random.choice(doors)

# Tell the user we've picked.
print("From Python: I've randomly selected a door to put the car behind.")
````

## Your first choice

Okay, so here's your first of two choices as the contestant: pick a door that you think the car might be behind.

Just to be clear, in the cell above Python randomly selected a door to put the car behind (implying there are goats behind the other two doors.)
You shouldn't have looked at what Python chose - it's stored in the `car` variable.
Now you're being asked to guess what door Python might have chosen, based on no information really - it's just a guess.
You should store that guess in the `pick` variable.

````python
# An example of picking the blue door.
pick = 'blue'
````

## Twist

At this point, the car has been randomly placed behind one of the doors and you've selected the door you think it might be behind.
Next the host throws in a twist: they open one of the doors that you haven't picked and shows you there is a goat behind it.
They then ask you whether you want to stick with your current pick or do you want to change to the other door that they didn't open.
Run the cell below to simulate the host doing this.

````python
# Pick a random door that was neither picked by the contestant nor has the car behind it.
show = random.choice([door for door in doors if door != car and door != pick])

# Figure out which door was not opened or picked.
notopen = [door for door in doors if door != pick and door != show][0]

# Tell the user which door was opened by the host.
from IPython.display import Markdown, display
display(Markdown(f"""
#### Host:
> *You chose the {pick} door. The producers have told me I can reveal to you that there is a goat behind the {show} door.
My question to you now is this: do you want to stick with the {pick} door or do you want to switch to the {notopen} door?*
"""))
````
